```python
import mindspore as ms
import mindspore.ops as ops
import numpy as np
import time

ms.set_context(device_target="CPU")

A = ms.Tensor(np.ones((2048, 2048), np.float16))
B = ms.Tensor(np.ones((2048, 1), np.float16))
```

```python
C = ops.MatMul()(A, B)
sum = 0.0
for _ in range(10000):
    start = time.time()
    C = ops.MatMul()(A, B)
    end = time.time()
    sum += (end - start) * 1000
print("MindSpore CPU Average Time:", sum / 10000, "ms")
```
